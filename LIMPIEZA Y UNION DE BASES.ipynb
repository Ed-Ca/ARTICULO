```python

def crear_punt_global(tabla, columnas):
    valores_array = array([col(columna) for columna in columnas])
    no_nulos = filter(valores_array, lambda dato: dato.isNotNull())

    # promedios = (
    #     aggregate(no_nulos, lit(0.0), lambda acc, x: acc+x) /
    #     coalesce(size(no_nulos), lit(1))
    # )

    promedios = when(size(no_nulos) == 0, lit(0.0)).otherwise(aggregate(no_nulos, lit(0.0), lambda acc, x: acc+x) / size(no_nulos))
    return tabla.withColumn("PUNT_GLOBAL", promedios)

```
